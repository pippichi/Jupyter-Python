```python
import requests
import re
import xlwt
import time

url = 'https://tieba.baidu.com/f?ie=utf-8&kw=胶原蛋白&fr=search'
proxypool_url = 'http://127.0.0.1:5555/random'
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.92 Safari/537.36"
}

def get_random_proxy():
    return requests.get(proxypool_url).text.strip()

def crawl(url,proxy):
    proxies = {'http':'http://'+proxy}
    return requests.get(url,proxies=proxies,headers=headers).text

def parse_one_page(html):
    pattern = re.compile('<li class=" j_thread_list clearfix".*?<div class="threadlist_title pull_left j_th_tit ">.*?<a.*?class="j_th_tit ">(.*?)</a>',re.S)
    items = re.findall(pattern,html)
    return items

def target(url):
    proxy = get_random_proxy()
    html = crawl(url,proxy)
    item = parse_one_page(html)
    return item
```

```python
target(url)
```
