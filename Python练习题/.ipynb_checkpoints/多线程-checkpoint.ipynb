```python
def thread_job2():
    print('T1 start')
    for i in range(10):
        time.sleep(0.1)
    print('T1 finish')
```

```python
def thread_job3():
    print('T2 start')
    print('T2 finish')
```

### Queue

```python
from queue import Queue
```

```python
def job(l,q):
    for i in range(len(l)):
        l[i] = l[i]**2
    q.put(l)
```

```python
def multithreading():
    q = Queue()
    threads = []
    data = [[1,2,3],[2,3,4],[3,4,5],[5,6,7]]
    for i in range(4):
        t = threading.Thread(target=job,args=(data[i],q))
        t.start()
        threads.append(t)
        
    for thread in threads:
        thread.join()
    result = []
    for _ in range(4):
        result.append(q.get())
    print(result)
```

```python
multithreading()
```

```python
from queue import Queue

def job(l,q):
    for i in range(len(l)):
        l[i] = l[i]**2
    q.put(l)
    
def multi():
    q = Queue()
    data = [[1,2],[2,3],[4,5],[5,6]]
    threads = []
    for i in range(len(data)):
        t = threading.Thread(target=job,args=(data[i],q))
        t.start()
        threads.append(t)
    result = []
    for t in threads:
        t.join()
    for _ in range(len(data)):
        result.append(q.get())
    return result
```

```python
multi()
```

### python的多线程不一定会有效率

```python
import copy

def job(l,q):
    res = sum(l)
    q.put(res)
    
def multi(l):
    q = Queue()
    threads,res = [],[]
    for i in range(4):
        t = threading.Thread(target=job,args=(copy.copy(l),q),name='T%i'%i)
        t.start()
        threads.append(t)
    [t.join() for t in threads]
    total = 0
    for _ in range(4):
        total += q.get()
    print(total)
    
def normal(l):
    total = sum(l)
    print(total)
    
l = list(range(10000000))
s_t = time.time()
normal(l*4)
print(time.time()-s_t)
s_t = time.time()
multi(l)
print(time.time()-s_t)
```

### 同时处理共享数据

```python
def job1():
    global A,lock
    lock.acquire()
    for i in range(10):
        A += 1 
        print('job1',A)
    lock.release()
        
def job2():
    global A,lock
    lock.acquire()
    for i in range(10):
        A += 10
        print('job2',A)
    lock.release()
```

```python
lock = threading.Lock()
A = 0
t1 = threading.Thread(target=job1)
t2 = threading.Thread(target=job2)
t1.start()
t2.start()
t1.join()
t2.join()
```

```python
from concurrent.futures import ThreadPoolExecutor
```

```python
def sayhello(a):
    global lock
    lock.acquire()
    print('hello:',a)
    lock.release()
    time.sleep(2)
def main():
    lock = threading.Lock()
    seed = list(range(10000))
    start = time.time()
    
    with ThreadPoolExecutor(1000) as executor:
        for i in seed:
            executor.submit(sayhello,i)
    end = time.time()
    print(end-start)
    start2 = time.time()
    
    with ThreadPoolExecutor(10) as executor:
        executor.map(sayhello,seed)
    
    end2 = time.time()
    print(end2-start2)
main()
```
