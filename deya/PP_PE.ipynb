```python
import pandas as pd 
import numpy as np
```

```python
PE_excel = pd.read_excel("爬虫-排号和排产.xlsx", sheet_name="PE排号", index_col=False)
PP_excel = pd.read_excel("爬虫-排号和排产.xlsx", sheet_name="PP排号", index_col=False)
```

```python
def get_name_category(excel, type_name):
    temp = excel.loc[:, ["型号", "类别"]]
    if type_name == "PP":
        temp.insert(len(list(temp.columns)), "type", np.zeros(len(temp)))
    elif type_name == "PE":
        temp.insert(len(list(temp.columns)), "type", np.ones(len(temp)))
    else:
        print("类型名错误")
        return 
    dict_name = {
        "型号": "specifications_name",
        "类别": "category"
    }
    temp.rename(columns=dict_name, inplace=True)
    return temp.dropna(axis=0, how="any")
```

```python
PE_temp = get_name_category(PE_excel, "PE")
PP_temp = get_name_category(PP_excel, "PP")
```
