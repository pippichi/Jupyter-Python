```python
def delta_n_date(date, n):
    """
    获取该日期前后几日的日期
    :param date: 日期字符串，格式为yyyy-mm-dd
    :param n: 整形数字，正数表示之后，负数表示之前
    :return: 日期字符串，格式为yyyy-mm-dd
    """
    offset = datetime.timedelta(days=n)
    result_date = (datetime.datetime.strptime(date, '%Y-%m-%d') + offset).strftime('%Y-%m-%d')

    return result_date
```

```python
def last_date_of_this_week(date):
    """
    获取该日期当周最后一天（周日）的日期
    :param date:日期字符串，格式为yyyy-mm-dd
    :return:日期字符串，格式为yyyy-mm-dd
    """
    # 首先获取当天是周几（返回值从0开始，周日是6）
    weekday = datetime.datetime.strptime(date, '%Y-%m-%d').weekday()
    # 接着获取(6-weekday)日后的日期
    result_date = delta_n_date(date, 4-weekday)

    return result_date
```

```python
class Result:
    result = {}
    def __init__(self):
        self.result.clear()
    def setResult(self, x, y):
        self.result[x] = y
        return self.result
    def getResult(self):
        return self.result

def move_and_loop(n):
    number=10**n
    result= Result()
    for each_num in range(number):
        if each_num not in result.getResult():
            value=f"{each_num:0{n}}"
            digits=[]
            while value not in digits:
                digits.append(value)
                first="".join(sorted(value,reverse=True)).ljust(n,'0')
                second="".join(sorted(value))
                value=f"{int(first)-int(second):0{n}}"
            index=digits.index(value)
            i=0
            while i<index:
                result.setResult(int(digits[i]), (index-i,len(digits)-index))
                i=i+1
            while i<len(digits):
                result.setResult(int(digits[i]), (0,len(digits)-index))
                i=i+1
    return result
```

```python
# 删除指定类型的文件
import os
import shutil

def delFunc(path):
    if os.path.isfile(path):
        try:
            os.remove(path)
        except Exception:
            path = path.replace("/", "\\")
            cmd = 'del "'+ path + '" /F'
            print(cmd)
            os.system(cmd)
            print(path)
    else:
        for o in os.listdir(path):
            copyPath = path
            copyPath = copyPath + "/" + o
            delFunc(copyPath)

def func(path):
    if os.path.isfile(path):
        return
    os.chdir(path)
    for o in os.listdir(os.curdir):
        copyPath = path
        copyPath = copyPath + "/" + o
        curFile = os.path.abspath(copyPath)
        if o == ".git":
            delFunc(curFile)
            shutil.rmtree(curFile)
            print(curFile)
        else:
            func(curFile)

func("e:/work/zk")
```

```python
def _check_if_exists(collectors, collector):
    flag = False
    collector_len = len(collector)
    for c in collectors:
        if len(c) != collector_len:
            continue
        i = 0
        while i != collector_len:
            if c[i] != collector[i]:
                break
            i += 1
        if i == collector_len:
            flag = True
            break
    return flag
def backtrace(data, collectors, collector, num, index):
    _collector = deepcopy(collector)
    if index == num or index == len(data):
        _collector = sorted(_collector)
        if _check_if_exists(collectors, _collector) is False:
            collectors.append(_collector)
        return
    for i in range(index, len(data)):
        if data[i] in _collector:
            continue
        _collector.append(data[i])
        backtrace(data, collectors, _collector, num, index + 1)
        _collector.pop()
```
