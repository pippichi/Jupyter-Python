```python
import requests
import re

url = 'http://guba.eastmoney.com/list,601318_{page}.html'
proxypool_url = 'http://127.0.0.1:5555/random'
headers = {
    'Accept': '*/*',
    'Accept-Encoding': 'gzip, deflate',
    'Accept-Language': 'en-US,en;q=0.9,zh-CN;q=0.8,zh;q=0.7',
    'Connection': 'keep-alive',
    'Host': 'emres.dfcfw.com',
    'Referer': 'http://guba.eastmoney.com/list,601318_1.html',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36',
}

def get_random_proxy():
    return requests.get(proxypool_url).text.strip()

def crawl(url,proxy):
    proxies = {'http':'http://'+proxy}
    return requests.get(url,proxies=proxies,headers=headers).text

def parse_one_page(html):
    pattern = re.compile('<div class="articleh.*?<a href.*?title="(.*?)">',re.S)
    items = re.findall(pattern,html)
    return items

def target(url):
    proxy = get_random_proxy()
    html = crawl(url,proxy)
    item = parse_one_page(html)
    return item
```
