```python
vt_st_rulefillna_summary = pd.read_excel(r'.\vt_mm_rulefillna_summary.xls',index_col=0)
vt_st_rulefillna_summary
```

```python
date = vt_st_rulefillna_summary.iloc[:,0].values
inclination = vt_st_rulefillna_summary.iloc[:,1].values
opening = vt_st_rulefillna_summary.iloc[:,2].values
closing = vt_st_rulefillna_summary.iloc[:,3].values
highest = vt_st_rulefillna_summary.iloc[:,4].values
lowest = vt_st_rulefillna_summary.iloc[:,5].values
rise_fall_rate = vt_st_rulefillna_summary.iloc[:,6].values
rise_fall_value = vt_st_rulefillna_summary.iloc[:,7].values
turnover = vt_st_rulefillna_summary.iloc[:,8].values
bussiness_volume = vt_st_rulefillna_summary.iloc[:,9].values
amplitude = vt_st_rulefillna_summary.iloc[:,10].values
turnover_rate = vt_st_rulefillna_summary.iloc[:,11].values
# index = np.array(range(1,10))
index = np.array(range(len(vt_st_rulefillna_summary)-1))
```

```python
plt.figure(figsize=(20,10))
plt.title('Inclination VS Others',fontsize=30)
plt.ylabel('inclination',fontsize=25)
plt.xlabel('("1" equals "2019-1-2","459" equals "2020-4-4")',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.scatter(index,inclination[:459],s=60,alpha=0.5,label='inclination')
plt.plot(index,opening[1:460]+0.5,label='opening')
plt.plot(index,closing[1:460]+0.5,label='closing')
plt.plot(index,highest[1:460]+0.5,label='highest')
plt.plot(index,lowest[1:460]+0.5,label='lowest')
plt.plot(index,rise_fall_rate[1:460],label='rise and fall rate')
plt.plot(index,rise_fall_value[1:460],label='rise and fall value')
plt.plot(index,turnover[1:460],label='turnover')
plt.plot(index,bussiness_volume[1:460],label='business volume')
plt.plot(index,amplitude[1:460],label='amplitude')
plt.plot(index,turnover_rate[1:460],label='turnover rate')

plt.legend()
```
