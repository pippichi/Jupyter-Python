```python
import numpy as np
import pandas as pd
```

```python
summary = pd.read_excel('.\summary.xls',index_col=False)
```

```python
nrows = len(summary)
ncols = summary.shape[1]
summary
```

```python
df = pd.DataFrame(summary)
```

```python
#均值填补
for i in range(1,ncols):
    nan_judgement = np.any(df.iloc[:,i].isnull())
    if (nan_judgement):
        rows = np.sum(np.array(df.iloc[:,i].isnull()==False))
        mean1 = np.sum(df.iloc[:,i]) / rows
        df.iloc[:,i] = df.iloc[:,i].fillna(value=mean1)
df
```

```python
#自定义逻辑填补
for i in range(1,ncols):
    nan_judgement = np.any(df.iloc[:,i].isnull())
    if (nan_judgement):
        for j in range(nrows):
            if(np.isnan(df.iloc[j,i])):
                if j==0:
                    df.iloc[j,i] = df.iloc[j+1,i] - (df.iloc[j+2,i]-df.iloc[j+1,i])
                elif (df.iloc[j-1,i]!=0 and df.iloc[j-2,i]!=0):
                    df.iloc[j,i] = df.iloc[j-1,i] + (df.iloc[j-1,i] - df.iloc[j-2,i])
df
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
column_names = df.columns.values
time_stamp = df.iloc[:,0]
time_stamp = pd.DataFrame(time_stamp)
time_stamp
```

```python
#to_numpy
data = np.array(df.iloc[:,1:])
```

```python
#kill_low_variance
threshold = VarianceThreshold(threshold=0.0)
res = threshold.fit_transform(data)
```

```python
for i in range(len(res)):
    if(res[i][0]>0.5):
        res[i][0] = 1
    else:
        res[i][0] = 0
res[:,0]
```

```python
res.shape
```

```python
#mean_01 or selfrule_01
for i in [5,6]:
    for j in range(len(res)):
        if(res[j][i]>=0):
            res[j][i] = 1
        else:
            res[j][i] = 0
res[:,[5,6]]
```

```python
#mean or selfrule
temp_res = res.copy()
for i in [1,2,3,4,7,8,9,10]:
    for j in range(1,len(res)):
        if((temp_res[j][i]-temp_res[j-1][i])>=0):
            res[j][i] = 1
        else:
            res[j][i] = 0
for i in range(1,11):
    res[0][i] = 0
```

```python
#subtract with mean or selfrule
temp_res = res.copy()
for i in [1,2,3,4,7,8,9,10]:
    for j in range(1,len(res)):
        res[j][i] = temp_res[j][i]-temp_res[j-1][i]
for i in range(1,11):
    res[0][i] = 0
```

```python
res
```

```python
res_to_pandas = pd.DataFrame(res,columns=column_names[1:-1])
res_total = pd.merge(time_stamp,res_to_pandas,on=time_stamp.index)
res_total = res_total.iloc[:,1:]
```

```python
res_total.to_excel('./vt_rulefillna_sub_summary.xls')
```

```python
#MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
minMaxScaler = MinMaxScaler()
minMaxScaler.fit(res)
print(minMaxScaler.scale_)
minMax_res = minMaxScaler.transform(res)
print(minMax_res[:5,:])
```

```python
#StandardScaler
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
standardScaler.fit(res)
print(standardScaler.scale_)
standard_res = standardScaler.transform(res)
print(standard_res[:5,:])
```

```python
#to pandas and merge
res_temp = standard_res
res_to_pandas = pd.DataFrame(res_temp,columns=column_names[1:-1])
print(res_to_pandas)
res_total = pd.merge(time_stamp,res_to_pandas,on=time_stamp.index)
res_total = res_total.iloc[:,1:]
```
