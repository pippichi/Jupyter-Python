```python
import xlrd
import xlwt
import numpy as np
import jieba
import codecs
from xlutils.copy import copy
```

```python
def load_words(address):
    words = []
    with codecs.open(address,'r','utf-8') as f:
        for each in f.readlines():
            each = each.strip('\n')
            each = each.strip('\r')
            words.append(each)
    return words
```

```python
stop_words = load_words('data/stop.txt')
pos_words = load_words('data/pos.txt')
neg_words = load_words('data/neg.txt')
```

```python
readbook = xlrd.open_workbook(r'.\original_comments.xls')
sheet = readbook.sheet_by_name('work_sheet1')
nrows = sheet.nrows
nrows
```

```python
text = []
for i in range(1,nrows):
    text.append(sheet.cell(i,1).value)
len(text)
```

```python
#replacement
replaced_result = []
for i in range(len(text)):
    replace_text = []
    sentence = jieba.cut(text[i],False)
    for word in sentence:
        if word in stop_words and word!=' ':
            continue
        else:
            if word in neg_words and word!=' ':
                replace_text.append("差")
            elif word in pos_words and word!=' ':
                replace_text.append("好")
            else:
                replace_text.append(word)
    res = "".join(replace_text)
    replaced_result.append(res)
replaced_result
```

```python
from urllib import request
import urllib
from urllib import parse
import json
from bs4 import BeautifulSoup
import requests
```

```python
Request_URL="http://fanyi.youdao.com/translate?smartresult=dict&smartresult=rule"
```

```python
headers = ("User-Agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36")
proxypool_url = 'http://127.0.0.1:5555/random'
def get_random_proxy():
    return requests.get(proxypool_url).text.strip()
```

```python
def translate(mandarin):
    form_data = {}
    form_data['i'] = mandarin
    form_data['from'] = 'AUTO'
    form_data['to'] = 'AUTO'
    form_data['smartresult'] = 'dict'
    form_data['doctype'] = 'json'
    form_data['version'] = '2.1'
    form_data['keyfrom'] = 'fanyi.web'
    form_data['action'] = 'FY_BY_CLICKBUTTION'
    form_data['typoResult'] = 'false'
    proxy = get_random_proxy()
    data = parse.urlencode(form_data).encode('utf-8')
    proxies = request.ProxyHandler({'http': proxy})
    opener = request.build_opener(proxies,request.HTTPHandler)
    opener.addheaders = [headers]
    request.install_opener(opener)
    response = request.urlopen(Request_URL,data)
    html = response.read().decode('utf-8')
    translate_results = json.loads(html)
    translate_result = translate_results["translateResult"][0][0]['tgt']
    return translate_result
```

```python
english_dict = []
EPOCH_SIZE = 10
BATCH_SIZE = 230
start = 0
for epoch in range(EPOCH_SIZE):
    end = start + BATCH_SIZE
    for i in range(start,end):
        english_dict.append(translate(text[i]))
    start += BATCH_SIZE
english_dict
```
