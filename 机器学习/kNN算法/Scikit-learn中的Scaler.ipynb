```python
import numpy as np
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X[:10,:]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=666)
```

### scikit-learn中的StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardScaler = StandardScaler()
```

```python
standardScaler.fit(X_train)
```

```python
standardScaler.mean_# 由于mean变量是由用户传进去的变量计算出来的，用户可以随时从外面进行查询，对于这种变量使用变量名后加_
```

```python
standardScaler.scale_ #输出方差,描述数据分布范围
```

```python
X_train = standardScaler.transform(X_train)
```

```python
X_train
```

```python
standardScaler.fit(X_test)
```

```python
X_test_standard = standardScaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn_clf.fit(X_train,y_train)
```

```python
knn_clf.score(X_test_standard,y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmaxScaler = MinMaxScaler()
```

```python
minmaxScaler.fit(X_train)
```

```python
X_train_MinMaxScaler = minmaxScaler.transform(X_train)
```

```python
X_train_MinMaxScaler
```

```python
minmaxScaler.fit(X_test)
```

```python
X_test_MinMaxScaler = minmaxScaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_clf.fit(X_train_MinMaxScaler,y_train)
```

```python
knn_clf.score(X_test_MinMaxScaler,y_test)
```
