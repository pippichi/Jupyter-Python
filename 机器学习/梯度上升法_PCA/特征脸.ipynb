# 特征脸

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_lfw_people
```

```python
face = fetch_lfw_people()
```

```python
face
```

```python
face.keys()
```

```python
face.data.shape
```

```python
face.images.shape #13233个图象，每个图像以62 x 47表示 62 x 47 = 2914
```

```python
random_indexes = np.random.permutation(len(face.data))
X = face.data[random_indexes]
```

```python
example_faces = X[:36,:]
example_faces.shape
```

```python
def plot_faces(faces):
    
    fig,axes = plt.subplots(6,6,figsize=(10,10),
                          subplot_kw={'xticks':[],'yticks':[]},
                          gridspec_kw=dict(hspace=0.2,wspace=0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(faces[i].reshape(62,47),cmap='bone')
    plt.show()

plot_faces(example_faces)
```

```python
face.target_names
```

```python
len(face.target_names)
```

```python
%%time
from sklearn.decomposition import PCA

pca = PCA(svd_solver='randomized')
pca.fit(X)
```

```python
pca.components_.shape
```

```python
plot_faces(pca.components_[:36,:]) #这个就是特征脸
```
