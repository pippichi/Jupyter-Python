# scikit-learn中的k_means

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

iris=datasets.load_iris()
x=iris.data[:,:2]
y=iris.target
```

```python
from sklearn.cluster import KMeans

km_clf=KMeans(n_clusters=3)
km_clf.fit(x)
```

```python
y_predict=km_clf.predict(x)
```

```python
plt.scatter(x[y_predict==0,0],x[y_predict==0,1])
plt.scatter(x[y_predict==1,0],x[y_predict==1,1])
plt.scatter(x[y_predict==2,0],x[y_predict==2,1])
plt.show()
```

```python
plt.scatter(x[y==0,0],x[y==0,1])
plt.scatter(x[y==1,0],x[y==1,1])
plt.scatter(x[y==2,0],x[y==2,1])
plt.show()
```
