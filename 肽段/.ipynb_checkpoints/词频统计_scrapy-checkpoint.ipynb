### 开始分析

```python
import numpy as np
import pandas as pd
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, TfidfTransformer
import codecs
```

```python
def loan_stopwords(path):
    stopwords = []
    with codecs.open(path, 'r', 'utf-8') as f:
        for each in f.readlines():
            each = each.strip('\n')
            each = each.strip('\r')
            stopwords.append(each)
    return stopwords
```

```python
stop_words = loan_stopwords('stopwords.txt')
```

```python
res = []
for i in total:
    i = i.strip()
    temp = jieba.cut(i, False)
    res_temp = []
    for j in temp:
        if j in stop_words:
            continue
        else:
            res_temp.append(j)
    res.append(" ".join(res_temp))
```

```python
res
```

```python
tfidf = TfidfVectorizer()
tfidf_model = tfidf.fit(res)
```

```python
temp = tfidf_model.vocabulary_
```

```python
temp.get("hand")
```

```python
sorted_list = sorted(temp.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted_list
```
