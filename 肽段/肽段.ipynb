```python
import requests
import re
import xlwt
import time

url = 'https://tieba.baidu.com/f?kw=胶原蛋白&ie=utf-8&pn={page}'
proxypool_url = 'http://127.0.0.1:5555/random'
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.92 Safari/537.36"
}

def get_random_proxy():
    return requests.get(proxypool_url).text.strip()

def crawl(url,proxy):
    proxies = {'http':'http://'+proxy}
    return requests.get(url,proxies=proxies,headers=headers).text

def parse_one_page(html):
    pattern = re.compile('<li class=" j_thread_list clearfix".*?<div class="threadlist_title pull_left j_th_tit ">.*?<a.*?class="j_th_tit ">(.*?)</a>',re.S)
    items = re.findall(pattern,html)
    return items

def target(url, page):
    proxy = get_random_proxy()
    html = crawl(url.format(page=page),proxy)
    item = parse_one_page(html)
    return item
```

```python
res = []
for i in range(0,10):
    temp = target(url, i*50)
    res.append(temp)
result = []
for i in res:
    for j in i:
        result.append(j)
```

```python
workbook = xlwt.Workbook(encoding='ascii')
table = workbook.add_sheet('work_sheet1')
table.write(0,0,'comments')
for i in range(len(result)):
    table.write(i+1,0,result[i])
```

```python
workbook.save('肽段.xls')
```
