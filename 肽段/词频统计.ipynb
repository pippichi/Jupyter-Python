```python
import numpy as np
import pandas as pd
import codecs
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
comments = pd.read_excel('肽段.xls',index_col=False)
```

```python
data = np.array(comments.iloc[:,0])
```

```python
data
```

```python
def loan_stopwords(path):
    stopwords = []
    with codecs.open(path, 'r', 'utf-8') as f:
        for each in f.readlines():
            each = each.strip('\n')
            each = each.strip('\r')
            stopwords.append(each)
    return stopwords
```

```python
stop_words = loan_stopwords('stopwords.txt')
```

```python
res = []
for i in data:
    temp = jieba.cut(i, False)
    res_temp = []
    for j in temp:
        if j in stop_words:
            continue
        else:
            res_temp.append(j)
    res.append(" ".join(res_temp))
```

```python
res
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf_model = tfidf.fit(res)
```

```python
tfidf_model.vocabulary_
```
